```python
import matplotlib.pyplot as plt
import keras_ocr

# Load the detector and recognizer models
pipeline = keras_ocr.pipeline.Pipeline()

# Provide the path to your text image
image_path = ''

# Load the image
images = [keras_ocr.tools.read(image_path)]

# Perform OCR
prediction_groups = pipeline.recognize(images)

# Display the image with the predicted bounding boxes and texts
fig, axs = plt.subplots(nrows=1, figsize=(15, 15))
keras_ocr.tools.drawAnnotations(image=images[0], predictions=prediction_groups[0], ax=axs)
plt.show()

# Extract the OCR text
ocr_text = []
for text_result in prediction_groups[0]:
    ocr_text.append(text_result[0])

# Print the OCR text
print("OCR Text:")
for text in ocr_text:
    print(text)
```
